```python
from collections import deque

# BFS implementation

# 1. 탐색 시작 노드를 큐에 삽입
# 2. 큐에서 노드를 꺼내 해당 노드의 인접 노드 중 방문하지 않은 노드가 있다면 모두 큐에 삽입
# 3. 2번의 과정을 더이상 반복할 수 없을 때까지 반복

queue = deque()

def BFS(graph, start):
    visited[start] = True
    print("visited : ", start)
    queue.append(start)
    
    while len(queue)>0:
        for i in graph[queue.popleft()]:
            if not visited[i]:
                print("visited : ", i)
                queue.append(i)
                visited[i] = True
    return None
    
    
graph = [
    [],
    [2, 3, 8],
    [1, 7],
    [1, 4, 5],
    [3, 5],
    [3, 4],
    [7],
    [2, 6, 8],
    [1, 7]
]
visited = [False]*9

BFS(graph,1)
```

```python
# 5-3 / 실전 문제 / 음료수 얼려 먹기
from collections import deque

n, m = map(int, input().split())

array = []

for i in range(n):
    array.append(list(map(int, input().split())))

visited = [[False]*m for _ in range(n)]

dx = [0, -1, 0, 1]
dy = [-1, 0, 1, 0]

queue = deque()

def baverage(array, x, y):
    global queue
    
    visited[x][y] = True
    queue.append((x,y))
    while queue:
        x, y = queue.popleft()
        for i in range(4):
            nx = x + dx[i]
            ny = y + dy[i]
            if nx < 0 or ny < 0 or nx >= n or ny >= m:
                continue
            if array[nx][ny] == 1 or visited[nx][ny]:
                continue
            visited[nx][ny] = True
            queue.append((nx,ny))
    
    return 1
        
count = 0

for i in range(n):
    for j in range(m):
        if array[i][j] == 0 and not visited[i][j]:
            count += baverage(array, i, j)
            
print(count)
```

```python
# 8-3 written by author, by using dfs, recursive function!

n, m = map(int, input().split())

graph = []
for i in range(n):
    graph.append(list(map(int, input().split())))

visited = [[False]*m for _ in range(n)]



def dfs(array, x, y):
    
    if x<0 or y <0 or x>=n or y>=m:
        return False
    
    if visited[x][y] == False:
        visited[x][y] = True
        dfs(array, x-1, y)
        dfs(array, x, y-1)
        dfs(array, x+1, y)
        dfs(array, x, y+1)
        return True
        
    return False


count = 0

for i in range(n):
    for j in range(m):
        if dfs(array, i, j):
            count+=1
            
            
print(count)

# 다시 한 번 dfs로 구현해보기! 재귀함수 사용해서.
```
