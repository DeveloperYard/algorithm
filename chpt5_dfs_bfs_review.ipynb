```python
from collections import deque

# BFS implementation

# 1. 탐색 시작 노드를 큐에 삽입
# 2. 큐에서 노드를 꺼내 해당 노드의 인접 노드 중 방문하지 않은 노드가 있다면 모두 큐에 삽입
# 3. 2번의 과정을 더이상 반복할 수 없을 때까지 반복

queue = deque()

def BFS(graph, start):
    visited[start] = True
    print("visited : ", start)
    queue.append(start)
    
    while len(queue)>0:
        for i in graph[queue.popleft()]:
            if not visited[i]:
                print("visited : ", i)
                queue.append(i)
                visited[i] = True
    return None
    
    
graph = [
    [],
    [2, 3, 8],
    [1, 7],
    [1, 4, 5],
    [3, 5],
    [3, 4],
    [7],
    [2, 6, 8],
    [1, 7]
]
visited = [False]*9

BFS(graph,1)
```

```python
# 5-3 / 실전 문제 / 음료수 얼려 먹기
from collections import deque

n, m = map(int, input().split())

array = []

for i in range(n):
    array.append(list(map(int, input().split())))

visited = [[False]*m for _ in range(n)]

dx = [0, -1, 0, 1]
dy = [-1, 0, 1, 0]

queue = deque()

def baverage(array, x, y):
    global queue
    
    visited[x][y] = True
    queue.append((x,y))
    while queue:
        x, y = queue.popleft()
        for i in range(4):
            nx = x + dx[i]
            ny = y + dy[i]
            if nx < 0 or ny < 0 or nx >= n or ny >= m:
                continue
            if array[nx][ny] == 1 or visited[nx][ny]:
                continue
            visited[nx][ny] = True
            queue.append((nx,ny))
    
    return 1
        
count = 0

for i in range(n):
    for j in range(m):
        if array[i][j] == 0 and not visited[i][j]:
            count += baverage(array, i, j)
            
print(count)
```

```python
# 8-4 / 실전 문제 / 미로 탈출 written by me!

from collections import deque

queue = deque()

n, m = map(int, input().split())
array = []

value = [[-1]*m for _ in range(n)]

for i in range (n):
    array.append(list(input()))

for i in range(n):
    for j in range(m):
        array[i][j] = int(array[i][j])
    
x, y = 0, 0
dx = [0, -1, 0, 1]
dy = [-1, 0, 1, 0]

queue.append((x, y))
value[x][y] = 1
while queue:
    x, y = queue.popleft()
    
    for i in range(4):
        nx = x + dx[i]
        ny = y + dy[i]
        
        # 첫 번째 조건문 : 범위를 벗어나거나 괴물이 있어 이동할 수 없는 경우
        
        if nx < 0 or ny < 0 or nx >= n or ny >= m or array[nx][ny] == 0:
            continue
            
        # 두 번째 조건문 : 이동할 수 있는 곳이나, 현재 노드에서 이동한 값보다 큰 경우(다른 먼 경로로 돌아온 경우)
        
        if value[nx][ny] != -1 and value[nx][ny] > value[x][y]+1:
            value[nx][ny] = value[x][y] + 1
        
        # 세 번째 조건문 : 방문하지 않은 노드인 경우
        
        elif value[nx][ny] == -1:
            value[nx][ny] = value[x][y] + 1
        
        else: continue
        queue.append((nx, ny))


print(value[n-1][m-1])

# solve!
```

```python
# 5-4 / written by author!

from collections import deque

n, m = map(int, input().split())

graph = []

for i in range(n):
    graph.append(list(map(int, input()))) # split을 안쓰면 그냥 나눠 주는 거였음! 내가 했던 것처럼 미련하게 하지 말자
    
dx = [-1, 1, 0, 0]
dy = [0, 0, -1, 1]

def bfs(x, y):
    queue = deque()
    queue.append((x, y))
    
    while queue:
        x, y = queue.popleft()
        for i in range(4):
            nx = x + dx[i]
            ny = y + dy[i]
            
            if nx < 0 or ny < 0 or nx >= n or ny >= m:
                continue
            if graph[nx][ny] == 0:
                continue
            if graph[nx][ny] == 1:
                graph[nx][ny] = graph[x][y]+1
                queue.append((nx, ny))
                
    return graph[n-1][m-1]

print(bfs(0, 0))
```
