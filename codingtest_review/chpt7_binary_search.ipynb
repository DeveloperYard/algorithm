```python
# 재귀 함수로 구현한 이진 탐색 소스 코드

def binary_search(arr, target, start, end):
    if start > end:
        return None
    
    mid = (start+end)//2
    
    if arr[mid] == target:
        return mid
    
    elif arr[mid] > target:
        return binary_search(arr, target, start, mid-1)
    else:
        return binary_search(arr, target, mid+1, end)
        
        
n, target = map(int, input().split())

arr = list(map(int, input().split()))

print(binary_search(arr, target, 0, len(arr)-1))
```

```python
# 반복문으로 구현한 이진 탐색 소스 코드

def binary_search(arr, target, start, end):
    while start <= end:
        mid = (start+end)//2
        
        if arr[mid] == target:
            return mid+1
        elif arr[mid] > target:
            end = mid-1
        else:
            start = mid+1
        
        
        
n, target = map(int, input().split())

arr = list(map(int, input().split()))

print(binary_search(arr, target, 0, len(arr)-1))
```

```python
# actual 3 ) 부품 찾기

def b_search(arr, n, start, end):
    if start > end:
        return "no"
    
    mid = (start+end)//2
    
    if arr[mid] == n:
        return "yes"
    elif arr[mid] > n:
        return b_search(arr, n, start, mid-1)
    else:
        return b_search(arr, n, mid+1, end)
    

def findOption(arr, target):
    ans = []
    
    for o in target:
        ans.append(b_search(arr, o, 0, len(arr)-1))
    
    return ans

n = int(input())
options = list(map(int, input().split()))

options.sort()

m = int(input())
targetOptions = list(map(int, input().split()))

print(findOption(options, targetOptions))
    
```

```python
# actual 3 ) 떡볶이 떡
def addDiff(arr, n):
    ans = 0
    for i in arr:
        if i > n:
            ans += i-n
    return ans

def bin_search(arr, target, start, end):
    result = 0
    
    end = max(arr)
    
    while start <= end:
        mid = (start+end)//2
        
        if addDiff(arr, mid) >= target: # 적어도 M이 되는 높이의 최대값을 구해야 하므로!
            result = mid
            start = mid+1
        else:
            end = mid-1
            
    return result


def findMinHeight(n, m, ls):
    ans = 0
    ls.sort()
    
    ans = bin_search(ls, m, 0, n-1)
    
    
    return ans

n, m = map(int, input().split())
ls = list(map(int, input().split()))

print(findMinHeight(n, m, ls))
```
