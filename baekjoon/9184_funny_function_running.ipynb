```python
# 신나는 함수 실행

# if a<=0 or b<=0 or c<=0, then w(a,b,c) return: 1
# if a>20 or b>20 or c>20, then w(a,b,c) return: w(20, 20, 20)
# if a<b and b<c, then w(a,b,c) return w(a,b,c-1)+w(a,b-1,c-1)-w(a,b-1,c)

# otherwise it returns:
# w(a-1, b, c) + w(a-1, b-1, c)+w(a-1, b, c-1) - w(a-1,b-1,c-1)
# -50 <= a,b,c <= 50

twVal = w(20,20,20)

# 한 번만 계산하고 계속 써먹을 수 있을 줄 알았는데, 아예 여기서부터 호출 후 너무 깊게 감!


def w(a,b,c):
    global twVal
    if a<=0 or b<=0 or c<=0: 
        return 1
    
    # 아래의 경우들을 위 경우로 보내야 함 
    
    elif a>20 or b>20 or c>20: # 이 놈이 가장 문제임!, 여기서 호출지옥 걸림
        return twVal
    
    elif a<b and b<c: # 순서가 정해져 있는 경우
        return w(a,b,c-1)+w(a,b-1,c-1)-w(a,b-1,c)

    else: # 무순서한 경우 + 0<a,b,c<20
        
        return w(a-1, b, c) + w(a-1, b-1, c)+w(a-1, b, c-1) - w(a-1,b-1,c-1)
    
while True:
    a, b, c = map(int, input().split())
    if a==-1 and b==-1 and c==-1: break
        
    print(f"w({a}, {b}, {c}) = ",w(a,b,c))


    
```
