```python
from collections import deque

m, n = map(int, input().split())

graph = []

days = [[0]*m for _ in range(n)]

q = deque()

dx = [-1, 0, 1, 0]
dy = [0, -1, 0, 1]

for _ in range(n):
    graph.append(list(map(int, input().split())))

visited = [[False]*m for _ in range(n)]

for i in range(n):
    for j in range(m):
        if graph[i][j] == 1:
            q.append((i, j, 1))
            

while q:
    for i in range(len(q)):
        x, y, day = q.popleft()
        visited[x][y] = True
        
        for i in range(4):
            nx = x+dx[i]
            ny = y+dy[i]
            if nx < 0 or ny < 0 or nx >= n or ny >= m:
                continue
            if graph[nx][ny] > 0 or graph[nx][ny] == -1:
                continue
                
            if visited[nx][ny] == True:
                continue
            
            graph[nx][ny] = day+1
            q.append((nx, ny, day+1))
            
result = -1
for i in range(n):
    for j in range(m):
        if graph[i][j] > 0:
            if graph[i][j] > result:
                result = graph[i][j]
            
    
    
flag = 0
for i in range(n):
    for j in range(m):
        if graph[i][j] == 0:
            flag = 1
            break
    
if flag:
    print(-1)
else:
    print(result-1)
    
# solve!!!
```
