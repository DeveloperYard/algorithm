```python
from collections import deque

m, n = map(int, input().split())
graph = []

for _ in range(n):
    graph.append(list(map(int, input().split())))
    
tomato = []

steps = [(-1, 0), (1, 0), (0, -1), (0, 1)]

for i in range(n):
    for j in range(m):
        if graph[i][j] == 1:
            tomato.append((i, j))

print(tomato)
def bfs(x, y):
    q = deque()
    q.append((x, y))
    
    while q:
        x, y = q.popleft()
        for step in steps:
            dx, dy = step
            nx = x + dx
            ny = y + dy
            if nx < 0 or ny < 0 or nx >= n or ny >= m:
                continue
            if graph[nx][ny] == -1:
                continue
            if graph[nx][ny] != 0 and graph[nx][ny] < graph[x][y]+1:
                continue
            
            if graph[nx][ny] == 0:
                graph[nx][ny] = graph[x][y]+1
                q.append((nx, ny))
                
            elif graph[nx][ny] >= graph[x][y]+1:
                graph[nx][ny] = graph[x][y]+1
                q.append((nx, ny))
            
    
def maxTomato():
    for coord in tomato:
        x, y = coord
        bfs(x, y)
        
    maxVal = -1
    for i in range(n):
        if maxVal < max(graph[i]):
            maxVal = max(graph[i])

#     for i in range(n):
#         print(graph[i])
        
    flag = 0
    for i in range(n):
        for j in range(m):
            if graph[i][j] == 0:
                flag = 1
                break
    if flag == 1:
        return -1
    else:
        return maxVal-1

print(maxTomato())
        

# don't solve.. must be review!!
```
