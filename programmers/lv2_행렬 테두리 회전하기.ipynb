```python
from collections import deque

def solution(rows, columns, queries):
    answer = []
    
    arr = [[0]*columns for _ in range(rows)]
    
    n = 1
    for i in range(rows):
        for j in range(columns):
            arr[i][j] = n
            n += 1
        
    for i in range(len(queries)):
        y1, x1, y2, x2 = queries[i]
        x1 = x1-1
        y1 = y1-1
        x2 = x2-1
        y2 = y2-1
        
        tmpArr = deque()
        
        seqArr = []
        
        x1Tmp = x1
        for _ in range(x2-x1):
            seqArr.append((y1, x1Tmp))
            x1Tmp+=1
        
        y1Tmp = y1
        for _ in range(y2-y1):
            seqArr.append((y1Tmp, x2))
            y1Tmp+=1
        
        x2Tmp = x2
        for _ in range(x2-x1):
            seqArr.append((y2, x2Tmp))
            x2Tmp -= 1
        
        y2Tmp = y2
        for _ in range(y2-y1):
            seqArr.append((y2Tmp, x1))
            y2Tmp -= 1
            
            
        
        for i in range(len(seqArr)):
            y, x = seqArr[i]
            tmpArr.append(arr[y][x])
            
        answer.append(min(tmpArr)) # 여기 추가하는 과정을 넣었으니 이제 회전시켜서 원본 numbers에 대입하면 됨!
        tmpArr.rotate(1)
        for i in range(len(seqArr)):
            y, x = seqArr[i]
            arr[y][x] = tmpArr[i]
        
        for i in range(rows):
            print(arr[i])
        print(seqArr)
        print(tmpArr)
        print(answer)
    
    return answer

r, c = map(int, input().split())
qLen = int(input())

queries = []
for i in range(qLen):
    queries.append(list(map(int, input().split())))
    
    
print(solution(r, c, queries))

# solve!! 행과 열 즉, arr[y][x] 이런 식으로 표현할 지 정확히 정의할 것!
```
