```python
from collections import deque


def bfs(q, dungeons):

    val = []
    while q:
        tired, dunArr, cnt = q.popleft()
        val.append(cnt)
        for i in range(len(dungeons)):
#             print(dunArr)
            if i not in dunArr:
                if dungeons[i][0] > tired:
                    continue
                if dungeons[i][1] > tired:
                    continue
                
                tmpTired = tired - dungeons[i][1]
                tmpDun = dunArr[:]
                tmpDun.append(i)
                q.append((tmpTired, tmpDun, cnt+1))

#     print("val : ")
#     print(val)
    return max(val)
def solution(k, dungeons):
    answer = -1
    
    q = deque()
    
    for i in range(len(dungeons)):
        reqT = dungeons[i][0]
        conT = dungeons[i][1]
        if reqT > k or k < conT:
            continue
        conquerDun = []
        conquerDun.append(i)
#         print("conquerDun : ")
#         print(conquerDun)
        q.append((k - conT, conquerDun, 1))
        curVal = bfs(q, dungeons)
        if answer < curVal:
            answer = curVal
        print(q)
        
    
    return answer


k = 80
dungeons = [[80,20],[50,40],[30,10]]

print(solution(k, dungeons))
```
