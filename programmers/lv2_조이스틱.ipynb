```python
def mimDist(idx, string, name):
    right = 0
    left = 0
    i = idx
    while string[i] == name[i]:
        i+=1
        if i >= len(name):
            i %= len(name)
        right+=1
    
    i = idx
    while string[i] == name[i]:
        i-=1
        if i < 0:
            i += len(name)
        left+=1
        
    return right if right < (-1)*left else left

def solution(name):
    answer = 0
    curIdx = 0
    
    curString = 'A'*len(name)
    
    while curString != name:
        if curString[curIdx] != name[curIdx]:
            answer += (ord(name[curIdx]) - ord('A'))
            tmpString = list(curString)
            tmpString[curIdx] = name[curIdx]
            curString = "".join(tmpString)
            print(curString)
        else:
            # 현재 인덱스에서 바꿔야 할 문자 중 가까운 곳을 찾음
            moveIdx = mimDist(curIdx, curString, name)
            print(moveIdx)
            if moveIdx < 0:
                curIdx += moveIdx
                
                if curIdx < 0:
                    curIdx+=len(name)
                answer+=(-1)*moveIdx
            
            else:
                curIdx += moveIdx
                
                if curIdx >= len(name):
                    curIdx%=len(name)
                
                answer+=moveIdx
                
        
    return answer


name = input()

print(solution(name))
```
